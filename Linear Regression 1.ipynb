```python
Y = A[["PROFIT"]]
X = A[["RND"]]
```

```python
xtrain = X[0:35]
xtest = X[35:50]
```

```python
ytrain = Y[0:35]
ytest = Y[35:50]
```

```python
ytrain.shape
```

```python
sb.distplot(Y)
sb.distplot(ytrain)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3)
sb.distplot(Y)
sb.distplot(ytrain)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(xtrain,ytrain)
```

```python
pred = model.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(ytest,pred)
```

```python
b0 = model.intercept_
b1 = model.coef_
```

```python
q0 = float(input("Enter the amount of money you wish to invest in RND"))
print("Estimated profit will be around %.2f"%(b0+b1*q0))
```
